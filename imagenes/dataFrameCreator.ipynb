```python
import json

# cargar el JSON original
with open("Mancuernas.json", "r") as f:
    original_data = json.load(f)

# crear una lista para guardar los datos actualizados
updated_data = []

# iterar a través de cada objeto en el JSON original
for item in original_data:
    # crear un diccionario para guardar los datos actualizados del objeto
    updated_item = {}

    # agregar el titulo del objeto al diccionario actualizado
    updated_item["title"] = item["title"]

    # crear una lista para guardar los enlaces actualizados del objeto
    updated_links = []

    # iterar a través de cada enlace en el objeto
    for link in item["images"]:
        # agregar solo los enlaces al diccionario actualizado
        updated_links.append(link["original"])

    # agregar la lista de enlaces al diccionario actualizado
    updated_item["images"] = updated_links

    # agregar el diccionario actualizado a la lista de datos actualizados
    updated_data.append(updated_item)

# guardar los datos actualizados en un nuevo archivo JSON
with open("Mancuernas1.json", "w") as f:
    json.dump(updated_data, f)
```

```python
import json

# cargar el JSON desde el archivo mancuernas1.json
with open('mancuernas1.json', 'r') as f:
    data = json.load(f)

# iterar sobre cada elemento del JSON y eliminar todos los enlaces excepto el primero
for item in data:
    item['images'] = [item['images'][0]]

# escribir el JSON modificado de vuelta al archivo mancuernas1.json
with open('mancuernas1.json', 'w') as f:
    json.dump(data, f, indent=4)
```
